```python
# 빅데이터 분석기사 실기 대비 Pandas 라이브러리 실습
# Pandas 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기
# 데이터는 구글드라이브에 저장 후 마운트하기
emp = pd.read_csv("/content/drive/MyDrive/datasets/emp.csv")
```

```python
# emp에서 사원의 이름과 월급 출력
emp[['ename', 'sal']]
```

```python
# emp에서 사원의 사원번호, 이름, 월급, 직업 출력
emp[['empno', 'ename', 'sal', 'job']]
```

```python
# emp에서 월급이 3000인 사원들의 이름, 월급 출력
emp[['ename', 'sal']][emp['sal'] == 3000]
```

```python
# emp에서 직업이 SALESMAN인 사원들의 이름, 월급, 직업 출력
emp[['ename', 'sal', 'job']][emp['job'] == 'SALESMAN']
```

```python
# emp에서 월급이 1200 이상인 사원들의 이름, 월급, 직업 출력
emp[['ename', 'sal', 'job']][emp['sal'] >= 1200]
```

```python
# emp에서 부서번호가 20번인 사원들의 이름, 월급, 부서번호 출력
emp[['ename', 'sal', 'deptno']][emp['deptno'] == 20]
```

```python
# emp에서 월급이 1000 ~ 3000 사이인 사원들의 이름, 월급 출력
emp[['ename', 'sal']][emp['sal'].between(1000, 3000)]
```

```python
# emp에서 월급이 1000 ~ 3000 사이가 아닌 사원들의 이름, 월급 출력
emp[['ename', 'sal']][~ emp['sal'].between(1000, 3000)]
```

```python
# 한글 데이터 파일 인코딩해서 불러오기
emp20 = pd.read_csv("/content/drive/MyDrive/datasets/emp20.csv", encoding="euckr")
```

```python
# emp20에서 나이가 20대인 학생들의 이름, 나이 출력
emp20[['ename', 'age']][emp20['age'].between(20, 29)]
```

```python
# emp20에서 통신사가 kt, lg인 학생들의 이름, 통신사 출력
emp20[['ename', 'telecom']][emp20['telecom'].isin([ 'kt', 'lg' ])]
```

```python
# emp20에서 통신사가 kt, lg가 아닌 학생들의 이름, 통신사 출력
emp20[['ename', 'telecom']][~ emp20['telecom'].isin([ 'kt', 'lg' ])]
```

```python
# emp의 결측치 컬럼 확인
emp.columns[emp.isna().any()]
```

```python
# emp에서 커미션이 NULL인 사원들의 이름, 커미션 출력
emp[['ename', 'comm']][emp['comm'].isnull()]
```

```python
# emp에서 이름이 'A'로 시작하는 사원들의 이름 출력
emp[['ename']][emp['ename'].apply(lambda x : x[0] =='A')]
```

```python
# emp에서 이름의 두번째 철자가 'M'인 사원의 이름 출력
emp[['ename']][emp['ename'].apply(lambda x : x[1] =='M')]
```

```python
# emp에서 이름이 'T'로 끝나는 사원의 이름, 월급 출력
emp[['ename']][emp['ename'].apply(lambda x : x[-1] =='T')]
```

```python
# emp20에서 성씨가 '김'씨인 학생들의 이름, 나이 출력
emp20[['ename', 'age']][emp20['ename'].apply(lambda x : x[0] == '김')]
```

```python
# emp20에서 성씨가 '김'씨가 아닌 학생들의 이름, 나이 출력
emp20[['ename', 'age']][emp20['ename'].apply(lambda x : x[0] != '김')]
```

```python
# emp에서 사원들의 이름, 월급을 월급이 낮은 사원부터 출력
emp[['ename', 'sal']].sort_values('sal', ascending=True)
```

```python
# emp에서 사원들의 이름, 월급, 직업을 월급이 높은 사원부터 출력
emp[['ename', 'sal', 'job']].sort_values('sal', ascending=False)
```

```python
# emp 데이터 정보 확인
emp.info()
```

```python
# emp의 hiredate를 날짜형으로 변환
emp['hiredate'] = pd.to_datetime(emp['hiredate'])
emp.info()
```

```python
# 81년 11월 17일에 입사한 사원들의 이름, 입사일 출력
emp[['ename', 'hiredate']][emp['hiredate'] == '1981-11-17']
```

```python
# 81년도에 입사한 사원들의 이름, 입사일 출력
emp[['ename', 'hiredate']][emp['hiredate'].dt.year == 1981]
```

```python
# emp에서 부서번호를 중복을 제거해서 출력
emp['deptno'].unique()
```

```python
# emp에서 부서번호가 20번인 사원들의 직업을 중복 제거해서 출력
emp['job'][emp['deptno'] == 20].unique()
```

```python
# emp20에서 통신사가 sk, kt인 학생들의 나이를 중복 제거해서 출력
emp20['age'][emp20['telecom'].isin(['sk', 'kt'])].unique()
```

```python
# emp20에서 통신사를 대문자로 출력
emp20['telecom'].str.upper()
```

```python
# emp에서 소문자 이름과 월급 출력
pd.concat([emp['ename'].str.lower(), emp['sal']], axis=1)
```

```python
# emp의 이름을 첫글자는 대문자로 나머지는 소문자로 출력
emp['ename'].str.capitalize()
```

```python
# emp에서 연도(hire_year) 컬럼을 추가하고 연도별 평균 월급 출력
emp['hire_year'] = emp['hiredate'].dt.year
emp.groupby('hire_year')['sal'].mean().reset_index()
```

```python
# emp에서 직업별 인원수 출력
emp.groupby('job')['empno'].count().reset_index()
```

```python
# emp에서 직업별 인원수가 3명 이상인 직업만 출력
emp2 = emp.groupby('job')['empno'].count().reset_index()
emp2[:][emp2['empno'] >= 3]
```

```python
# emp에서 부서번호별, 직업별로 총 월급을 출력
emp.groupby(['deptno', 'job'])['sal'].sum().reset_index()
```

```python
# 타이타닉의 Age의 결측치를 평균값으로 채우기
tit = pd.read_csv('/content/drive/MyDrive/datasets/titanic.csv')
tit['Age'] = tit['Age'].fillna(tit['Age'].median())
tit['Age'].value_counts()
```

```python
# 타이타닉의 Age의 결측치를 중앙값으로 채우기
tit = pd.read_csv('/content/drive/MyDrive/datasets/titanic.csv')
tit['Age'] = tit['Age'].fillna(tit['Age'].median())
tit['Age'].value_counts()
```

```python
# 타이타닉의 Age의 결측치를 최빈값으로 채우기
tit = pd.read_csv('/content/drive/MyDrive/datasets/titanic.csv')
tit['Age'] = tit['Age'].fillna(tit['Age'].mode().iloc[0])
tit['Age'].value_counts()
```
